```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(2)
```

```python
movies['cast'][0]
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head(2)
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head(2)
```

```python
#convert overview column in list
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head(2)
```

```python
#now wwe have to remove space 
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head(2)
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head(2)
```

```python
#now convert list into string
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df.head(2)
```

```python
new_df['tags'][0]
```

```python
#convert all into lower case
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

## Now we stem all words

```python
import nltk
```

```python

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

## Now we have to do text Vectorisation
## Bag of Words
in this techinc we concatinate all the tags of all movies. after that we pick most repeted words..now we distribute all the repeted words according to they have came into privious movie tag..in this technic we dont use stop words like(and,are,a,is..etc)..

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
#from this we get most used words
cv.get_feature_names()
```

```python
# now we have to find vector distance one movie to another movie
# distance is inversily perposnal to similarity it means that if movies are similar to each other than distance is less otherwise distance distance is high
```

```python
## Now to have to find similarity 1= more similar, 0 = less similar
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
similarity
```

```python
similarity[0]
```

```python

# now we define a function---> when we search a movie it recomend 5 similar movie
# firts we have to sort all the distance and when we search a movie it gives nearest 5 movie name
```

```python
# how to find index
new_df[new_df['title']=='Avatar'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
#this is my recommender system
recommend("Pirates of the Caribbean: At World's End")
```
